```python
import numpy as np
import os

destino = "../olis_sells/data/raw"
file_name = "olist_order_payments_dataset.csv"
# complete_file_name = os.path.join(destino, "ollis_order_payments_dataset.csv")
# print("borrame", complete_file_name)

np_original_array = np.genfromtxt(os.path.join(destino, file_name), delimiter=",", skip_header=1)
print('np original array: \n', np_original_array)
original_shape = np_original_array.shape
print('np_original_array shape: \n', original_shape)
```

### Lets add a ones vector lettings "nan values" intact

```python
changed_original = np_original_array + np.ones(original_shape)*1
print('changed original array: \n', changed_original)
```

### Lets add 2 vector full of 2's inside our original_array cleaned version 

```python

# Limpiar el arreglo para eliminar los nan
cleaned_arr = np.isfinite(np_original_array)
cleaned_arr_shape = cleaned_arr.shape
print('original_shape: \n', original_shape, ' \n cleaned_arr_shape: \n', cleaned_arr_shape)

filtered_array = ~np.isnan(np_original_array)
# Filtrar el arreglo original con np.where (el 0 susttuye los varoles)
filtered_array = np.where(np.isnan(np_original_array), 0, np.ones(original_shape))
# print('filtered_array', filtered_array)

added_array = np.add(np_original_array, filtered_array*10)
print('added array', filtered_array)
```

### Crear una matriz aleatoria del tamaño del original array
